# First Steps

##### Pandas allows us to store our data in tabular, multi-dimensional objects with familiar features like rows, columns, and headers. You will now import that pandas package by using the command below. This allows us to call funcitons from pandas with pd instead of pandas.  

```python
import pandas as pd
```

# Using the pd.read_csv() function

##### pd.read_csv reads the tabular data from a Comma Seperated Values (csv) file into a dataframe object we'll define as JB. To create our dataframe object we'll type in exactly what is written below.  

```python
JB=pd.read_csv("CountyHealthData_2014-2015 (1).csv")
```

# Pulling Out The Data

##### To select the certain columns we want we list the names of the columns inside brackets (JB[["Year", "County", "Adult obesity"]]). To select the certain rows we want we put the numbers of the rows we want in brackets (JB[751:1068]). Look below to see exactly how it is done. 

```python
JB[["Year", "County", "Adult obesity"]][751:1068]
```

```python
JB[["Year", "County", "Food environment index"]][751:1068]
```

# Saving Subset

##### In order to save the subsets of data you will type in what is below, but instead of GAobesity and GAfoodenvironmentindex you will type in whatver you want to name it. 

```python
GAobesity_subset=JB[["Year", "County", "Adult obesity"]][751:1068].copy()
```

```python
GAfoodenvironmentindex_subset=JB[["Year", "County", "Food environment index"]][751:1068].copy()
```

##### Now if you ever want to use your subsets you just type in the name of it and _subset and you can see below.

```python
GAobesity_subset
```

```python
GAfoodenvironmentindex_subset
```

# Exporting New Subset

##### Now that we are done creating our subset we can export it as a .csv file in our working directory. To do this we type in exactly what is written below, but instead of GAobesity or GAfoodenvironmentindext type in what you named your subsets.  

```python
GAobesity_subset.to_csv("GAobesity_subset.csv", index=False)
```

```python
GAfoodenvironmentindex_subset.to_csv("GAfoodenvironmentindex_subset.csv", index=False)
```
