# First Steps

##### Pandas allows us to store our data in tabular, multi-dimensional objects with familiar features like rows, columns, and headers. You will now import that pandas package by using the command below. This allows us to call funcitons from pandas with pd instead of pandas.  

```python
import pandas as pd
```

# Using the pd.read_csv() function

##### pd.read_csv reads the tabular data from a Comma Seperated Values (csv) file into a dataframe object we'll define as John. To create our dataframe object we'll type in exactly what is written below.  

```python
John=pd.read_csv("CountyHealthData_2014-2015 (1).csv")
```

# Pulling Out The Data

##### To select the certain columns we want we list the names of the columns inside brackets (John[["State", "Year", "County"]]). To select the certain rows we want we put the numbers of the rows we want in brackets (John[751:1068]). Look below to see exactly how it is done. 

```python
John[["State", "Year", "County", "Premature death", "Adult smoking", "Adult obesity", "Food environment index", "Excessive drinking", "Teen births", "Primary care physicians"]][751:1068]
```

# Saving Subset

##### In order to save the subset of data you will type in what is below, but instead of Gahealthdata you will type in whatver you want to name it. 

```python
Gahealthdata_subset=John[John["State"]=="GA"].copy()
```

##### Now if you ever want to use your subset you just type in the name of it and _subset as you can see below.

```python
Gahealthdata_subset
```

# Exporting New Subset

##### Now that we are done creating our subset we can export it as a .csv file in our working directory. To do this we type in exactly what is written below, but instead of Gahealthdata type in what you named the subset.  

```python
Gahealthdata_subset.to_csv("Gahealthdata.csv",index=False)
```
